### Ex E5 ⭐⭐ — Fonction extract()
Encapsulez tout dans une fonction `extract(q1_path, q2_path, clients_path, produits_path) -> dict`
qui retourne un dictionnaire avec les 4 DataFrames harmonisés et consolidés.
La fonction doit afficher des logs à chaque étape.

```python
# etape1 création de la fonction extract avec ses différents paramètres puis chargement des fichiers
def extract(q1_path, q2_path, clients_path, produits_path):
    print("Chargement des fichiers")
    df_commandes1 = pd.read_csv(q1_path)
    df_commandes2 = pd.read_csv(q2_path)
    df_clients = pd.read_csv(clients_path)
    df_produits = pd.read_csv(produits_path)
    print("Chargement des fichiers terminés")

    # etape 2 ) harmonisation des données pour s'assurer que les 2 tableaux ont la meme struture
    print("harmonisation des données ")
    # les 2 tableaux seront similaires
    df_commandes2 = df_commandes2.rename(columns={'moyen_paiement': 'mode_paiement'})
    # ajout de la colonne devise manquante sur le tableau 1
    df_commandes1['devise'] = 0

    # etape 3 ) concaténation ou consolidation
    print("consolidation en cours ")
    df_commandes1['source'] = 'Q1_2024'
    df_commandes2['source'] = 'Q2_2024'

    # fusion de 2 tableaux avec concat et ignore index pour avoir une numeration logique de l'index
    df_commandes = pd.concat([df_commandes1, df_commandes2], ignore_index=True)

    # création du dictionnaire
    print("création du dictionnaire")
    resultats = {
        'commandes' : df_commandes,
        'clients' : df_clients,
        'produits' : df_produits
    }
    
    print("Extraction effectuée avec succès")
    return resultats

#  execution du code
# chemin opérationnel
chemin_q1 = r'C:\Users\eraze\Downloads\ETL\data\commandes_Q1.csv'
chemin_q2 = r'C:\Users\eraze\Downloads\ETL\data\commandes_Q2.csv'
chemin_clients = r'C:\Users\eraze\Downloads\ETL\data\clients.csv'
chemin_produits = r'C:\Users\eraze\Downloads\ETL\data\produits.csv'

# On appelle la fonction et on stocke le résultat
mes_donnees = extract(chemin_q1, chemin_q2, chemin_clients, chemin_produits)

# affichage des résultats 
print("Aperçu des commandes")
display(mes_donnees['commandes'].head(5))

print("aperçu des clients")
display(mes_donnees['clients'].head(5))

print("aperçu des produits")
display(mes_donnees['produits'].head(5))
```

### Ex T7 ⭐⭐ — Feature Engineering
Créez les colonnes suivantes :
1. `tranche_montant` : `<100€` / `100-500€` / `500-2000€` / `>2000€` avec `pd.cut()`
2. `saison` : depuis le mois (Hiver/Printemps/Été/Automne)
3. `est_annulee` : booléen True si statut == 'annulée'
4. `ca_perdu` : montant_total si commande annulée, sinon 0
5. `delai_inscription_commande` : nb de jours entre date_inscription du client et date_commande

Affichez `value_counts()` pour les colonnes catégorielles créées.

```python
# création des limites de pd.cut
bins = [0,100,500,2000,float('inf')]
# création des labels de pd.cut
labels = ['<100€', '100-500€','500-2000€','>2000€']

# création de la colonne tranche_montant avec pd cut 
df_commandes['tranche_montant'] = pd.cut(df_commandes['montant_total'],bins = bins , labels = labels )

# création de la fonction pour les saisons et de la colonne saison
def saisons(mois):
    if mois in [12,1,2]:
        return 'hiver'
    elif mois in [3,4,5]:
        return 'printemps'
    elif mois in [6,7,8]:
        return 'été'
    else:
        return 'automne'

df_commandes['saison'] = df_commandes['date_commande'].dt.month.apply(saisons)

# création de la colonne est_annulee
df_commandes['est_annulee'] = df_commandes['statut'] == 'annulée'

# création de la colonne ca_perdu
df_commandes['ca_perdu'] = 0.0
df_commandes.loc[df_commandes['est_annulee'] == True , 'ca_perdu'] = df_commandes['montant_total']
# df_commandes

# création de la colonne delai 
# 1. On force Pandas à lire ces deux colonnes comme de vraies dates mathématiques
df_commandes['date_commande'] = pd.to_datetime(df_commandes['date_commande'])
df_commandes['date_inscription'] = pd.to_datetime(df_commandes['date_inscription'])

# 2. Maintenant que ce sont deux vraies dates, la soustraction va fonctionner !
df_commandes['delai_inscription_commande'] = (df_commandes['date_commande'] - df_commandes['date_inscription']).dt.days

# Affichage pour vérifier le résultat
display(df_commandes)


# affichage des values_counts
print(df_commandes['tranche_montant'].value_counts())
print(df_commandes['saison'].value_counts())
print(df_commandes['est_annulee'].value_counts())



    
    

```
