### Ex E5 â­â­ â Fonction extract()
Encapsulez tout dans une fonction `extract(q1_path, q2_path, clients_path, produits_path) -> dict`
qui retourne un dictionnaire avec les 4 DataFrames harmonisÃĐs et consolidÃĐs.
La fonction doit afficher des logs Ã  chaque ÃĐtape.

```python
# etape1 crÃĐation de la fonction extract avec ses diffÃĐrents paramÃĻtres puis chargement des fichiers
def extract(q1_path, q2_path, clients_path, produits_path):
    print("Chargement des fichiers")
    df_commandes1 = pd.read_csv(q1_path)
    df_commandes2 = pd.read_csv(q2_path)
    df_clients = pd.read_csv(clients_path)
    df_produits = pd.read_csv(produits_path)
    print("Chargement des fichiers terminÃĐs")

    # etape 2 ) harmonisation des donnÃĐes pour s'assurer que les 2 tableaux ont la meme struture
    print("harmonisation des donnÃĐes ")
    # les 2 tableaux seront similaires
    df_commandes2 = df_commandes2.rename(columns={'moyen_paiement': 'mode_paiement'})
    # ajout de la colonne devise manquante sur le tableau 1
    df_commandes1['devise'] = "EUR"

    # etape 3 ) concatÃĐnation ou consolidation
    print("consolidation en cours ")
    df_commandes1['source'] = 'Q1_2024'
    df_commandes2['source'] = 'Q2_2024'

    # fusion de 2 tableaux avec concat et ignore index pour avoir une numeration logique de l'index
    df_commandes = pd.concat([df_commandes1, df_commandes2], ignore_index=True)

    # crÃĐation du dictionnaire
    print("crÃĐation du dictionnaire")
    resultats = {
        'commandes' : df_commandes,
        'clients' : df_clients,
        'produits' : df_produits
    }
    
    print("Extraction effectuÃĐe avec succÃĻs")
    return resultats

#  execution du code
# chemin opÃĐrationnel
chemin_q1 = r'C:\Users\eraze\Desktop\ETL\data\commandes_Q1.csv'
chemin_q2 = r'C:\Users\eraze\Desktop\ETL\data\commandes_Q2.csv'
chemin_clients = r'C:\Users\eraze\Desktop\ETL\data\clients.csv'
chemin_produits = r'C:\Users\eraze\Desktop\ETL\data\produits.csv'

# On appelle la fonction et on stocke le rÃĐsultat
mes_donnees = extract(chemin_q1, chemin_q2, chemin_clients, chemin_produits)

# affichage des rÃĐsultats 
print("AperÃ§u des commandes")
display(mes_donnees['commandes'].head(5))

print("aperÃ§u des clients")
display(mes_donnees['clients'].head(5))

print("aperÃ§u des produits")
display(mes_donnees['produits'].head(5))
```

### Ex T7 â­â­ â Feature Engineering
CrÃĐez les colonnes suivantes :
1. `tranche_montant` : `<100âŽ` / `100-500âŽ` / `500-2000âŽ` / `>2000âŽ` avec `pd.cut()`
2. `saison` : depuis le mois (Hiver/Printemps/ÃtÃĐ/Automne)
3. `est_annulee` : boolÃĐen True si statut == 'annulÃĐe'

```python
# crÃĐation des limites de pd.cut
bins = [0,100,500,2000,float('inf')]
# crÃĐation des labels de pd.cut
labels = ['<100âŽ', '100-500âŽ','500-2000âŽ','>2000âŽ']

# crÃĐation de la colonne tranche_montant avec pd cut 
df_commandes['tranche_montant'] = pd.cut(df_commandes['montant_total'],bins = bins , labels = labels )

# crÃĐation de la fonction pour les saisons et de la colonne saison
def saisons(mois):
    if mois in [12,1,2]:
        return 'hiver'
    elif mois in [3,4,5]:
        return 'printemps'
    elif mois in [6,7,8]:
        return 'ÃĐtÃĐ'
    else:
        return 'automne'

df_commandes['saison'] = df_commandes['date_commande'].dt.month.apply(saisons)

# crÃĐation de la colonne est_annulee
df_commandes['est_annulee'] = df_commandes['statut'] == 'annulÃĐe'

# crÃĐation de la colonne ca_perdu
df_commandes['ca_perdu'] = 0.0
df_commandes.loc[df_commandes['est_annulee'] == True , 'ca_perdu'] = df_commandes['montant_total']
# df_commandes

# crÃĐation de la colonne delai 
# 1. On force Pandas Ã  lire ces deux colonnes comme de vraies dates mathÃĐmatiques
df_commandes['date_commande'] = pd.to_datetime(df_commandes['date_commande'])
df_commandes['date_inscription'] = pd.to_datetime(df_commandes['date_inscription'])

# 2. Maintenant que ce sont deux vraies dates, la soustraction va fonctionner !
df_commandes['delai_inscription_commande'] = (df_commandes['date_commande'] - df_commandes['date_inscription']).dt.days

# Affichage pour vÃĐrifier le rÃĐsultat
display(df_commandes)


# affichage des values_counts
print(df_commandes['tranche_montant'].value_counts())
print(df_commandes['saison'].value_counts())
print(df_commandes['est_annulee'].value_counts())



    
    

```

### Ex T8 â­â­â­ â Validation qualitÃĐ
Ãcris une fonction `valider_pipeline(df) -> bool` qui :
1. VÃĐrifie qu'il n'y a plus de NaN sur les colonnes critiques
2. VÃĐrifie qu'il n'y a plus de doublons
3. VÃĐrifie que tous les prix sont > 0
4. VÃĐrifie que toutes les quantitÃĐs sont > 0
5. VÃĐrifie que montant_total == quantite Ã prix_unitaire (Ã  1 centime prÃĻs)
6. Retourne True si tout est OK, False sinon avec les erreurs dÃĐtaillÃĐes

```python
def valider_pipeline(df)-> bool:
# verification plus de Nan
    if df[['date_commande','statut','montant_total']].isna().sum().sum() > 0:
        return ("NaN dÃĐtectÃĐ")
   
# verification des doublons 
    if df.duplicated().sum() > 0:
        return("doublons dÃĐtectÃĐs")
        
# verification que tous les prix sont >0
    if (df["prix_unitaire"] <= 0).any() :
        return ("les prix nÃĐgatifs dÃĐtectÃĐs")
        
# verification quantite >=0
    if (df["quantite"] <= 0).any():
        return ("quantitÃĐs nÃĐgatives dÃĐtectÃĐes")
       
# verification montant total ==  qte*px
    if (df['montant_total'] != df['quantite'] * df['prix_unitaire']).any():
        return ("le montant total est correct")

    return True 
    
    
    
```
